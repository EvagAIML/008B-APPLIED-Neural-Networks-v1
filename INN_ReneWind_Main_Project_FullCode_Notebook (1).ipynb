### **Problem Statement**

**Business Context**

Renewable energy sources play an increasingly important role in the global energy mix as efforts to reduce the environmental impact of energy production increase.

Among all renewable energy alternatives, wind energy is one of the most developed technologies worldwide. The U.S. Department of Energy has created a guide to achieving operational efficiency using predictive maintenance practices.

Predictive maintenance uses sensor information and analytical methods to measure and predict degradation and future component capability. The principle behind predictive maintenance is that failure patterns are predictable. If component failure can be accurately predicted and the component replaced before it fails, the overall costs of operation and maintenance can be significantly reduced.

Sensors fitted across different machines involved in energy generation collect data related to various environmental factors such as temperature, humidity, and wind speed, as well as additional features related to different parts of the wind turbine, including the gearbox, tower, blades, and brake.

**Objective**

“ReneWind” is a company working to improve the machinery and processes involved in wind energy production using machine learning. The company has collected data on generator failures of wind turbines using sensors. The data shared is a ciphered version of the actual data due to confidentiality requirements (as the type of data collected varies across companies). The dataset includes 40 predictors, with 20,000 observations in the training set and 5,000 in the test set.

The goal is to build, tune, and compare various classification models to **determine** the best-performing model for identifying failures. This will allow generators to be repaired before they fail, thereby reducing overall maintenance costs.

**The model’s predictions can be interpreted as follows:**

True Positives (TP): Failures correctly predicted by the model, leading to repair costs.

False Negatives (FN): Actual failures not detected by the model, leading to replacement costs.
  
False Positives (FP): Incorrect detections where no failure occurs, leading to inspection costs.

It is known that the cost of repairing a generator is significantly lower than the cost of replacing it, and the cost of inspection is lower than the cost of repair.

**The target variable uses the following encoding:**

“1” = Failure

“0” = No failure

**Data Dictionary**

The data provided is a transformed version of the original sensor data.

Train.csv: Used for training and tuning the models.

Test.csv: Used for testing the performance of the final selected model.

Both datasets consist of 40 predictor variables and 1 target variable.
